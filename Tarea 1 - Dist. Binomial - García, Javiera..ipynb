**Distribución de probabilidad binomial**

*La distribución binomial describe el número de éxitos al realizar n experimentos independientes entre sí.*

P(x) = $
\left(
\begin{array}{ll}
n\\
x
\end{array}
\right)$
$p^{x}$
$q^{n-x}$

Donde: 

$\left(
\begin{array}{ll}
n\\
x
\end{array}
\right)$
$=$
$\frac{n!}{x!(n-x)!}$



x = Número de éxitos   
n = Número de ensayos/experimentos       
p = Probabilidad de éxito       
q = probabilidad de fracaso = 1 - p

**Distribución de probabilidad de Poisson**

*Poisson describe el número de sucesos de un determinado evento en un período de tiempo.*

F(x) = P(X=x) = (e^-/lambda)(lambda^x) / x!

$\frac{e^{\lambda}}$
${x!(n-x)!}$

Donde:




```python
mu =  4.8 # parametro de forma 
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='green', lw=5, alpha=0.5)
plt.title('Distribución Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
